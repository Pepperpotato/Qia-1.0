```python
# 导入文件
filename = 'chaos.txt' 
```

```python
# 文件读写
mytext = open(filename,encoding = 'utf-8').read()
```

```python
mytext
```

```python
# 引入词云模块
from wordcloud import WordCloud
```

```python
# 已经完成词云生成，未显示
mycloud = WordCloud().generate(mytext) 
```

```python
mycloud
```

```python
# 告诉jupyter notebook 请把文件显示在笔记本中
%matplotlib inline 
```

```python
# 读入python默认的绘图工具
import matplotlib.pyplot as plt
```

```python
# 使用plt中的imshow功能  ，并设置编码格式    
plt.imshow(mycloud,interpolation='bilinear')
```
